```python
import pandas as pd
import pickle

metrics_df1 = pickle.load(open("metrics_df1.pkl", "rb"))#0-1
metrics_df2 = pickle.load(open("metrics_df2.pkl", "rb"))#2-3
metrics_df3 = pickle.load(open("metrics_df3.pkl", "rb"))#4-5
metrics_df4 = pickle.load(open("metrics_df4.pkl", "rb"))#6-7
metrics_df5 = pickle.load(open("metrics_df5.pkl", "rb"))#8-9
metrics_df6 = pickle.load(open("metrics_df6.pkl", "rb"))#10-11
metrics_df7 = pickle.load(open("metrics_df7.pkl", "rb"))#12-13
metrics_df8 = pickle.load(open("metrics_df8.pkl", "rb"))#14-15
metrics_df9 = pickle.load(open("metrics_df9.pkl", "rb"))#16-17
metrics_df10 = pickle.load(open("metrics_df10.pkl", "rb"))#18-19


#combine all metrics dataframes
metrics_df = pd.concat([metrics_df1, metrics_df2, metrics_df3, metrics_df4, metrics_df5, metrics_df6, metrics_df7, metrics_df8, metrics_df9, metrics_df10], ignore_index=True)
metrics_df.sort_values(by='RMSE', ascending=True, inplace=True)
metrics_df

#The bests are the data set 'upsampled_df_smote.csv' AND 'df_oversamling.csv'
```

```python
import matplotlib.pyplot as plt

# Create a new figure
plt.figure(figsize=(10, 6))

# Plot RMSE with lines and points
plt.plot(range(metrics_df.shape[0]), metrics_df['RMSE'], marker='o', linestyle='-', label='RMSE')

# Plot MAE with lines and points
plt.plot(range(metrics_df.shape[0]), metrics_df['MAE'], marker='s', linestyle='-', label='MAE')

# Plot MSE with lines and points
plt.plot(range(metrics_df.shape[0]), metrics_df['MSE'], marker='^', linestyle='-', label='MSE')

# Annotate points with their index
for i in range(metrics_df.shape[0]):
    plt.annotate(i, (i, metrics_df['RMSE'].iloc[i]))
    plt.annotate(i, (i, metrics_df['MAE'].iloc[i]))
    plt.annotate(i, (i, metrics_df['MSE'].iloc[i]))

# Set title and labels
plt.title('Error Metrics for Different Models')
plt.xlabel('Model Index')
plt.ylabel('Error Value')

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
